### Handling Missing Data

There are multiple ways to handle missing data
*keep the missing data as is
*Remove an entire row or column with missing data
*Remove rows with a lot of missing data
*Impute missing numerical data with a 0 or a substitute like the
average, mode etc.
*Resolve the missing data based on your domain expertise

### Keeping missing data

You can still perform calculations if you choose to keep missing data even if there are NaN present

### Standard Deviation

The standard deviation is a measure of the spread of a dataset from the mean

* Values at least 3 standard deviations away from the mean are considered outliers
(meant for normally distibuted or bell shaped, data)
* The threshold of 3 std, can be changed to 2 or 4+ depending on the data

```python
# get the mean, and standard deviation for the grade column

mean = np.mean(df["Grade"])
```

```python
sd = np.std(df["Grade"])
```

```python
mean, sd
```

```python
[grade for grade in df["Grade"] if (grade < mean - 3*sd) or (grade > mean + 2*sd)]
```

```python
df["Grade"].sort_values()
```

### Creating New Columns

After cleaning data types and issues, you may still not
have the exact data that you need, so you can create new columns
from existing data to aid your analysis.

* Numeric columns: calculating percentages, applying conditional calculations etc.
* Datetime columns: extracting datetime components, applying datetime calculations, etc.
* Text columns: extracting text, splitting into multiple columns, finding patterns, etc.

```python
# Creating numeric columns

# calculating percentages

# set up 2 columns one wth the numerator and one with ythe
# denominator valies and divide them (mutiply 100 )


# calculating based on a condition

# use np.where() to create a new column based on a logical condition

```

```python
groceries = pd.read_excel("Data/Groceries.xlsx")
groceries.head()
```

```python
groceries["New Column"] = round(groceries["Price_Dollars"] * 1.05, 2)
```

```python
groceries.head()
```

```python
# what percent of each row of the Inventory

groceries["Total Inventory"] = groceries["Inventory"].sum()
```

```python
groceries["Percent Inventory"] = round(groceries["Inventory"] / groceries["Total Inventory"] * 100, 2)
```

```python
groceries.head()
```

```python
# flag any item having item of 50 and below as low inventory

groceries["Low Inventory"] = np.where(groceries["Inventory"] < 50, 'Low Inventory', '')
```

```python
groceries.tail()
```

### Extracting Datetime components

Use dt.component to extract a component from a datime value(day, month, etc.)

```python
# dt.date (date without time component), 
# dt.year (Year), 
# dt.month (numeric month 1 to 12),
# dt.day (day of the month)
# dt.dayofweek (numeric weekday (mon=0, sun=6)),
# dt.time (time without date component),
# dt.hour (hour (0-23)),
# dt.minute (minute (0-59))
# dt.second (second(0-59))

# use pd.to_timedelta() to add or subtract a particular timeframe
```

```python
groceries.head()
```

```python
# extract some of these dates

# time

groceries["Last_Updated_Time"] = groceries["Last_Updated"].dt.time
```

```python
# extract the day of the week
groceries["Shipment_Date_DOW"] = groceries["Next_Scheduled_Shipment"].dt.dayofweek
```

```python
groceries.head()
```

```python
dow_mapping = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
```

```python
groceries["Shipment_Date_DOW"] = groceries["Shipment_Date_DOW"].map(dow_mapping)
```

```python
groceries.head()
```

```python
groceries["New Shipment Date"] = groceries["Next_Scheduled_Shipment"] + pd.to_timedelta(1, 'D')
```

```python
# add 1 day with coditionals

groceries["New Shipment Date"] = np.where(groceries["Category"] == 'Produce: Fruit', 
         groceries["Next_Scheduled_Shipment"] + pd.to_timedelta(1, 'D'),
         groceries["Next_Scheduled_Shipment"])
```

```python
groceries.head()
```
